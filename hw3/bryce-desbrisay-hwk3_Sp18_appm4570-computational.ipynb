## Problem 1

There is an urn that has 100 marbles in it: 2 are red, 70 are yellow, 13 are green and 15 are blue. Simulate data to imitate someone drawing a marble from the urn (replacing after each draw), with the color recorded at each draw (HINT: the ${\tt sample()}$ function in ${\tt R}$ will be useful here).

#### (a) Plot the fraction of times a yellow marble is drawn in the first 1000 draws. On the x-axis, include the number of draws, and on the y-axis, include the cumulative fraction of yellow draws. Why does this number fluctuate more when the number of draws is smaller?

```r
r = rep("r",2)
y = rep("y",70)
g = rep("g",13)
b = rep("b",15)

set.seed(123)

marbles = c(r,y,g,b)
draws = sample(marbles, 1000, replace=TRUE)

p = matrix(nrow=1000, ncol=1)

for (i in c(1:1000)) {
    temp = draws[1:i]
    p[i] = length(temp[temp == "y"])/i
}

plot(p, type="l", xlab="Num of Draws", ylab="Fraction of yellow")
abline(h=c(0.7))
```

## Problem 2

#### (a) Run a simulation that estimates the probability calculation in theoretical question 1 (b). That is, simulate the birthdays (you can use the numbers 1 through 365)  of 45 people many many times, and find the relative frequency of cases in which at least two people share a birthday (HINT: The ${\tt sample()}$ and ${\tt replicate()}$ functions in ${\tt R}$ will be useful here).

```r
set.seed(1470)
counts = matrix(ncol=1,nrow=1000)
total = 0;

for (k in c(1:1000)) {
    birthdays = sample(c(1:365), 45, replace=TRUE)
    count = 0;

    for (i in c(1:45)) {
        for (j in c(1:45)) {
            if (birthdays[j] == birthdays[i] && i != j) {
                count = count + 1;
            }
        }
    }
    
    counts[k] = count / 2
    total = total + (count / 2)
}

"Average shared birthdays: "
total / 1000

"Probability of two or more people sharing a birthday with 45 people: "
total / 1000 / 45

"Roughly 6% probability of having the same birthday as someone else in the room"

boxplot(counts)
```

#### (b) Run a simulation that estimates the probability calculation in theoretical question 1 (c).

```r
set.seed(730)
rolls = replicate(2000, sample(c(1:6), 3, replace=TRUE))
totals = matrix(ncol=1, nrow=2000)
p12 = matrix(ncol=1, nrow=2000)
p13 = matrix(ncol=1, nrow=2000)

for (i in c(1:2000)) {
    totals[i] = rolls[1,i] + rolls[2,i] + rolls[3,i]
    temp = totals[1:i]
    p12[i] = length(temp[temp == 12])/i
    p13[i] = length(temp[temp == 13])/i
}

plot(p12, type="l", ylab="Probability of 12")
abline(h=0.11)
"There appears to be roughly an 11% chance that the sum of three rolls equals 12"

plot(p13, type="l", ylab="Probability of 13")
abline(h=0.1)
"There appears to be roughly an 10% chance that the sum of three rolls equals 13"
```

#### (c) Run a simulation that estimates the probability calculation in theoretical question 4 (a).

```r
"6/6 * 1/6 * 1/6 * 1/6 * 1/6 = 0.0007716 = 154321/200000000 = Prob of Yahtzee"

set.seed(20984357)
yrolls = replicate(30000, sample(c(1:6), 5, replace=TRUE))
yahtzees = matrix(ncol=1, nrow=30000)
yp = matrix(ncol=1, nrow=30000)


for (i in c(1:30000)) {
    if (yrolls[1,i] == yrolls[2,i] && yrolls[1,i] == yrolls[3,i] && yrolls[1,i] == yrolls[4,i] && yrolls[1,i] == yrolls[5,i]) {
        yahtzees[i] = 1
    } else {
        yahtzees[i] = 0
    }
    ytemp = yahtzees[1:i]
    yp[i] = length(ytemp[ytemp == 1])/i
}

plot(yp, type="l")

"Estimated robability of a Yahtzee on the first turn: "
length(yahtzees[yahtzees == 1]) / 30000
```
