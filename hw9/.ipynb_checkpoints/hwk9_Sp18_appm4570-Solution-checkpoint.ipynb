## Problem 1

#### (a) Generate 10,000 samples of size $n_1 = 20$ from $X \sim N(8,2)$ and 10,000 samples of size $n_2 = 20$ from $Y\sim N(6,2)$. Use ${\tt set.seed(89)}$ so that the answers are reproducible. Assume $X$ and $Y$ are independent and $\alpha = 0.05$. What type of distribution does the statistic $s_x^2\big/s_y^2$ follow? (HINT: See unit 7 code.)


```r
x = matrix(nrow=1000, ncol=20)
y = matrix(nrow=1000, ncol=20)
xsd = matrix(nrow=1000, ncol=1)
ysd = matrix(nrow=1000, ncol=1)
vars = matrix(nrow=1000, ncol=1)
set.seed(89)

for (i in c(0:1000)) {
    x[i,] = rnorm(20, mean=8, sd=2)
    y[i,] = rnorm(20, mean=6, sd=2)
    xsd[i,] = sd(x[i,])
    ysd[i,] = sd(y[i,])
    vars[i,] = xsd[i,]^2 / ysd[i,]^2
}

"var x / var y follows an F distribution with a hump shifted to the left of the graph"
```

#### (b) Create a density histogram of the 10,000 values of $s_x^2\big/s_y^2$. Superimpose the theoretical sampling distribution of $s_x^2\big/s_y^2$ on the density histogram.

```r
hist(vars, density=TRUE, breaks=40)
curve(df(x, df1=20, df2=20)*100, add=TRUE)
```

## Problem 2

The manager of a web company is interested in predicting the number of megabytes downloaded, ${\tt megasd}$, by clients according to the number of minutes they are connected, ${\tt mconnected}$. The mamager randomly selects clients, records the data, and stores the (${\tt megasd, mconnected}$) pairs in the file ${\tt url.txt}$.

#### (a) Create a scatterplot of the data. Characterize the relationship between ${\tt megasd}$ and ${\tt mconnected}$.



```r
data = data.frame(read.table("./url.txt", sep="\t", header=TRUE))
plot(data$megasd, data$mconnected, type="p")

"This plot has characteristics of a linear relationship between megasd and mconnected"
```

#### (b)  Fit a regression line to the data. Superimpose the resulting line in the plot created in part (a).


```r
lmod = lm(mconnected ~ megasd, data=data)
plot(data$megasd, data$mconnected)
abline(lmod)
summary(lmod)
```

#### (c) Compute the covariance matrix of the estimated regression coefficients. 



```r
vcov(lmod)
```

#### (d) What is the standard error of $\widehat{\beta}_1$?



```r
summary(lmod)$coefficients[2,2]
```

#### (e) What is the covariance between $\widehat{\beta}_0$ and $\widehat{\beta}_1$?



```r
b1 = summary(lmod)$coefficients[1,1]
b0 = summary(lmod)$coefficients[2,1]

mean = mean(data$megasd)
paste("Cov(b0, b1) = ",
(-mean * sd(data$megasd)^2)/(sum((x - mean)^2)))
```

#### (f) Construct a 90% confidence interval for the slope of the regression line.


```r
newData = data.frame(megasd = mean)
predict(lmod, newData, interval="confidence", level=0.9)
"(9.285 to 9.845)"
```

#### (g) How much variability in ${\tt megasd}$ can be explained by ${\tt mconnected}$?

```r
paste(b0, "%")
```

#### (h) What assumptions need to be satisfied in order to use this SLR model for inferential purposes?

#### (i) Estimate the mean value of megabytes downloaded by clients spending 4, 8, and 12 minutes online.


```r
ytox = function(y) {
    return((y-b1)/b0)
}

(ytox(4) + ytox(8) + ytox(12))/3
```
