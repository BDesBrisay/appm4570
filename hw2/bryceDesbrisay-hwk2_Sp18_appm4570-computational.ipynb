Some claim that the final hours aboard the Titanic were marked by class warfare; other claim it was characterized by male chivalry.  Load the titanic data frame into R. #### (a) Determine the fraction of survivors from each passenger class.

```r
t = data.frame(read.table("./titanic3.csv", sep=",", header=TRUE))
t$X...pclass = factor(t$X...pclass)
t$survived = factor(t$survived)
ts = t[t$survived == 1,]
td = t[t$survived == 0,]

fc = t[t$X...pclass == 1,]
first = (nrow(fc[fc$survived == 1,]))/(nrow(fc))

sc = t[t$X...pclass == 2,]
second = (nrow(sc[sc$survived == 1,]))/(nrow(sc))

tc = t[t$X...pclass == 3,]
third = (nrow(tc[tc$survived == 1,]))/(nrow(tc))

first
second
third
```

#### (b) Compute the fraction of survivors according to class and gender. Did men in the first class or women in the third class have a higher survival rate?

```r
fm = (nrow(fc[(fc$survived == 1) & (fc$sex == 'male'),]))/(nrow(fc[fc$sex == 'male',]))
ff = (nrow(fc[(fc$survived == 1) & (fc$sex == 'female'),]))/(nrow(fc[fc$sex == 'female',]))

sm = (nrow(sc[(sc$survived == 1) & (sc$sex == 'male'),]))/(nrow(sc[sc$sex == 'male',]))
sf = (nrow(sc[(sc$survived == 1) & (sc$sex == 'female'),]))/(nrow(sc[sc$sex == 'female',]))

tm = (nrow(tc[(tc$survived == 1) & (tc$sex == 'male'),]))/(nrow(tc[tc$sex == 'male',]))
tf = (nrow(tc[(tc$survived == 1) & (tc$sex == 'female'),]))/(nrow(tc[tc$sex == 'female',]))

genderClass = data.frame(c(1,2,3), c(fm,sm,tm), c(ff,sf,tf))
names(genderClass) = c("Class", "Male", "Female")

genderClass

"Women all across the board, regardless of class, had a higher survival rate compared to Men"
```

#### (d) Were the median and mean ages for females who survived higher or lower than for females who did not survive? Report the median and mean ages as well as an appropriate measure of spread for each statistic.

```r
falive = ts[ts$sex == 'female',]
fdead = td[td$sex == 'female',]
falive

faages = na.omit(falive$age)
fdages = na.omit(fdead$age)

fsum = data.frame(
    c("Median", "Mean", "SD", "Var"), 
    c(median(faages),mean(faages),sd(faages), var(faages)), 
    c(median(fdages),mean(fdages),sd(fdages), var(fdages))
)
names(fsum) = c("Measure", "Alive", "Dead")

"Both the mean and median for ages were higher for the female survivors"
fsum
```

#### (e) Were the median and mean ages for males who survived higher or lower than for males who did not survive? Report the median and mean ages as well as an appropriate measure of spread for each statistic.

```r
malive = ts[ts$sex == 'male',]
mdead = td[td$sex == 'male',]

maages = na.omit(malive$age)
mdages = na.omit(mdead$age)

msum = data.frame(
    c("Median", "Mean", "SD", "Var"), 
    c(median(maages),mean(maages),sd(maages), var(maages)), 
    c(median(mdages),mean(mdages),sd(mdages), var(mdages))
)
names(msum) = c("Measure", "Alive", "Dead")

"Both the mean and median for ages were lower for the male survivors"
msum
```
